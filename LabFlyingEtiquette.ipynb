```python
# keep important data

clean_df = df[['Age', 'How tall are you?', 'Is it rude to ask someone to switch seats with you in order to be closer to friends?', 'Do you ever recline your seat when you fly?', 'Is it rude to wake a passenger up if you are trying to go to the bathroom?', 'Under normal circumstances, does a person who reclines their seat during a flight have any obligation to the person sitting behind them?', 'Given the opportunity, would you eliminate the possibility of reclining seats on planes entirely?']]

clean_df.head()
```

```python
# replace "under 5ft" to "5 ft"
# change to 5'0 for next step instead

clean_df = clean_df.replace('Under 5 ft.','''5'0"''')

# change 6'6" and above	to 6'6"
clean_df = clean_df.replace('''6'6" and above''', '''6'6"''')

clean_df
```

```python
# check values
clean_df['How tall are you?'].value_counts()
```

```python
# Create a new column, “Height (Inches)”, that converts the “How tall are you?” column in feet and inches to an integer of total inches.

# drop nulls 
clean_df.dropna(subset=['How tall are you?'], inplace=True)
clean_df
```

```python
# new column to convert

feet = clean_df['How tall are you?'].str.slice(stop=1)
inches = clean_df['How tall are you?'].str.slice(start=2, stop=-1)

clean_df['Height (Inches)'] = (pd.to_numeric(feet)*12) + (pd.to_numeric(inches))
clean_df['Height (Inches)']
```
